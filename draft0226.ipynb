```python
import numpy as np
import pandas as pd
import re,math
from sklearn import preprocessing
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
from pyecharts import options as opts
from pyecharts.charts import Graph



data1 = pd.read_csv(r'2022_02_23_07_33_27.csv',encoding='gbk')
data2 = pd.read_excel(r'选择你所在家乡城市 描述统计.xlsx')
data3 = pd.read_excel(r'学校以及学院 描述统计.xlsx')
```

```python
# 去除非二元问题
temp = data1.iloc[1:,23:].fillna(0)
tp1 = []
for index, row in temp.iterrows():
    temp1 = []
    for j in row:
        if j == 0:
            temp1.append(0)
        elif j == "0":
            temp1.append(0)
        elif j == "1":
            temp1.append(1)
        elif j == "是":
            temp1.append(1)
        elif j == "否":
            temp1.append(0)
        else:
            # print(j)
            temp1.append(999)
    tp1.append(temp1)
temp2 = pd.DataFrame(tp1, columns=temp.columns)
temp2
```

```python
# 填补空缺
temp3 = []
columns_f = []
for columns, line in temp2.iteritems():
    status_f = 0
    for j in line:
        if j == 999:
            status_f = 1
        else:
            pass
    if status_f == 0:
        temp3.append(line)
        columns_f.append(columns)
temp4 = pd.DataFrame(temp3,index=columns_f).T
temp4
```

```python
# 筛选点
nodes = []
origin_nodes = []
count = 0
for column, line in temp4.iteritems():
    try:
        c = column.split("-")[1]
    except:
        c = column
    ct = {"name": c, "symbolSize": line.sum()}
    #print(line, count)
    if (count >= 16 and count <= 22) or (count >= 41 and count <= 47):# (count>=6 and count<=10) or 
        nodes.append(ct)
        origin_nodes.append(line)
    count += 1
```

```python
# 做斯皮尔曼相关度分析
tpt = []
cpp = pd.DataFrame(origin_nodes).T.corr('spearman')
for column, line in pd.DataFrame(origin_nodes).T.corr('spearman').iteritems():
    tptt = []
    for j in line:
        if abs(j) >= 0.07:
            tptt.append(1)
        else:
            tptt.append(0)
    tpt.append(tptt)
temp5 = np.array(pd.DataFrame(tpt, columns=cpp.columns, index=cpp.index))
temp5
```

```python
# 聚合成图
links = []
for i,k in zip(nodes, temp5):
    for j,l in  zip(nodes, k):
        if l == 1:
            links.append({"source": i.get("name"), "target": j.get("name")})
c = (
    Graph()
    .add("", nodes, links, repulsion=8000)
    .set_global_opts(title_opts=opts.TitleOpts(title="兼职聚合图"))
    .render("聚合图5.html")
)
```
