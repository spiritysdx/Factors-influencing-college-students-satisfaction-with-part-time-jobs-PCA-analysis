```python
import json

from pyecharts import options as opts
from pyecharts.charts import PictorialBar

location = ["—线城市", "家乡城市", "学校所在城市", "其他城市"]
values = [163, 85, 185, 17]


with open("symbol.json", "r", encoding="utf-8") as f:
    symbols = json.load(f)


c = (
    PictorialBar()
    .add_xaxis(location)
    .add_yaxis(
        "",
        values,
        label_opts=opts.LabelOpts(is_show=False),
        symbol_size=22,
        symbol_repeat="fixed",
        symbol_offset=[0, -5],
        is_symbol_clip=True,
        symbol=symbols["boy"],
    )
    .reversal_axis()
    .set_global_opts(
        title_opts=opts.TitleOpts(title="城市"),
        xaxis_opts=opts.AxisOpts(is_show=False),
        yaxis_opts=opts.AxisOpts(
            axistick_opts=opts.AxisTickOpts(is_show=False),
            axisline_opts=opts.AxisLineOpts(
                linestyle_opts=opts.LineStyleOpts(opacity=0)
            ),
        ),
    )
    .render("城市.html")
)
```

```python
from pyecharts import options as opts
from pyecharts.charts import Funnel



a = ["一个月","一到二个月","二到四个月","四到六个月","大于六个月"]
b = [19, 67, 68, 20, 15]

c = (
    Funnel()
    .add(
        "月份",
        [list(z) for z in zip(a, b)],
        sort_="ascending",
        label_opts=opts.LabelOpts(position="inside"),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Funnel-Sort（ascending）"))
    .render("月份1.html")
)
```
